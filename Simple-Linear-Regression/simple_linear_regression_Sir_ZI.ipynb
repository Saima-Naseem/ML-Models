```python
# 1-Simple Linear Regression
# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
# convert data to a dataframe

df = pd.read_excel('/content/mango_data.xlsx')
```

```python
# visualize data
plt.xlabel('Weight', color='red')
plt.ylabel('Price', color='Green')
plt.scatter(df.Weight, df.Price)
```

```python
# creating object for model training
lr = linear_model.LinearRegression()
lr.fit(df[['Weight']],df.Price)
```

```python
# PREDICTION

lr.predict([[11]])
```

```python
# Checking score

lr.score(df[['Weight']],df.Price)
```

```python
# Check Coefficient value

lr.coef_
```

```python
# check Intercept Value

lr.intercept_
```

```python
# Predicted value by formula. y=wx+b (when the coefficient is multiplied by the (predicted)input value and + it with intercept value, you get the predicted value.)

predicted = 100. * 11 + 0.0
print(predicted)
```

```python
# visualize data
plt.xlabel('Weight', color='red')
plt.ylabel('Price', color='Green')
plt.scatter(df.Weight, df.Price)
plt.plot(df.Weight, lr.predict(df[['Weight']]), color='green')
```
