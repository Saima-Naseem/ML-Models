```python
# Import Pandas and create variables carrying dataframes
import pandas as pd
df1 = pd.read_csv('/content/train.csv')
df2 = pd.read_csv('/content/test.csv')

df1.head()
```

```python
# drop unnecessary features and finalize target variable
x = df1.drop(columns=["Unnamed: 0", "PassengerId", "Survived"])
y = df1["Survived"]
```

```python
# Train DecisionTree model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x, y)
```

```python
# Check Accuracy on training data
model.score(x,y)
```

```python
# drop unnecessary columns from testing data

x_test = df2.drop(columns=['Unnamed: 0', 'PassengerId'])
```

```python
# Dropping the y label column 'survived' from testing data

X_test = df2.drop(columns=["Unnamed: 0", "PassengerId", "Survived"])
y_test = df2["Survived"]
```

```python
# make prediction on testing data
y_pred = model.predict(X_test)
```

```python
# Check accuracy score on testing data

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
# check confusion metrics score

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# measure Precision, Recall & F1 score

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
